```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc, classification_report
import xgboost as xgb


# Initialize XGBoost Classifier

xgb_model = xgb.XGBClassifier(
    objective='binary:logistic', # For binary classification
    eval_metric='logloss',       # Evaluation metric during training
    use_label_encoder=False,     # Suppress deprecation warning
    n_estimators=200,            # Number of boosting rounds
    learning_rate=0.1,           # Step size shrinkage to prevent overfitting
    random_state=42              # For reproducibility
)

# Train the model
xgb_model.fit(X_train, y_train)

print("XGBoost model training complete!")
```

```python
y_pred_bc=xgb_model.predict(X_test)

y_prob_bc =xgb_model.predict_proba(X_test)[:, 1] # Probability of liver disease (class 1)

# Calculate Evaluation Metrics
accuracy = accuracy_score(y_test, y_pred_bc)
precision = precision_score(y_test, y_pred_bc)
recall = recall_score(y_test, y_pred_bc)
f1 = f1_score(y_test, y_pred_bc)
roc_auc = roc_auc_score(y_test, y_prob_bc)
conf_matrix = confusion_matrix(y_test, y_pred_bc)
fpr, tpr, thresholds = roc_curve(y_test, y_prob_bc) # For ROC plot

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC-ROC: {roc_auc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_bc))

print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
print("\n--- PERFORMING HYPERPARAMETER TUNING (GridSearchCV) FOR BREAST CANCER MODEL ---")

xgb_model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss', use_label_encoder=False, random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.9],
    'colsample_bytree': [0.7, 0.9]
}

grid_search_xgb_bc = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='roc_auc', # Optimize for AUC-ROC
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search_xgb_bc.fit(X_train, y_train)

best_xgb_model_bc = grid_search_xgb_bc.best_estimator_
best_params_bc = grid_search_xgb_bc.best_params_

print("\n--- Breast Cancer Hyperparameter Tuning Complete ---")
print(f"Best AUC-ROC Score during CV: {grid_search_xgb_bc.best_score_:.4f}")
print(f"Best Parameters Found: {best_params_bc}")
```

```python
print("\n--- EVALUATING THE BEST BREAST CANCER MODEL ---")

y_pred_bc = best_xgb_model_bc.predict(X_test)
y_prob_bc = best_xgb_model_bc.predict_proba(X_test)[:, 1] # Probability of Malignant (class 1)

accuracy_bc = accuracy_score(y_test, y_pred_bc)
precision_bc = precision_score(y_test, y_pred_bc)
recall_bc = recall_score(y_test, y_pred_bc)
f1_bc = f1_score(y_test, y_pred_bc)
roc_auc_bc = roc_auc_score(y_test, y_prob_bc)
conf_matrix_bc = confusion_matrix(y_test, y_pred_bc)
fpr_bc, tpr_bc, thresholds_bc = roc_curve(y_test, y_prob_bc) # For ROC plot

print(f"Accuracy: {accuracy_bc:.4f}")
print(f"Precision: {precision_bc:.4f}")
print(f"Recall: {recall_bc:.4f}")
print(f"F1-Score: {f1_bc:.4f}")
print(f"AUC-ROC: {roc_auc_bc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_bc))

print("\nConfusion Matrix:")
print(conf_matrix_bc)

# Visualize Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_bc, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Benign (0)', 'Predicted Malignant (1)'],
            yticklabels=['Actual Benign (0)', 'Actual Malignant (1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Breast Cancer Prediction (Tuned Model)')
plt.show()

# Plot ROC Curve
plt.figure(figsize=(7, 6))
plt.plot(fpr_bc, tpr_bc, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_bc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve (Tuned Model)')
plt.legend(loc="lower right")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc, classification_report
import xgboost as xgb
import joblib
import os
import sys # Import sys for stderr messages

# --- 1. Data Acquisition - Load both datasets ---

# a) Load scikit-learn's Breast Cancer dataset
print("Loading scikit-learn's Breast Cancer (Diagnostic) Dataset...")
cancer_sk = load_breast_cancer()
df_sk = pd.DataFrame(data=cancer_sk.data, columns=cancer_sk.feature_names)
df_sk['diagnosis'] = cancer_sk.target

# Map sklearn target to our convention: 0 for Benign, 1 for Malignant
df_sk['diagnosis'] = df_sk['diagnosis'].apply(lambda x: 1 if x == 0 else 0)
print("Scikit-learn dataset loaded. Target mapped: 1=Malignant, 0=Benign.")
print(f"Scikit-learn dataset shape: {df_sk.shape}")
print(df_sk['diagnosis'].value_counts())
print("\nScikit-learn feature names (example):", df_sk.columns.tolist()[:5], "...")


# b) Load the uploaded breastCancer.csv file
print("\nLoading breastCancer.csv from uploaded files...")
try:
    df_csv = pd.read_csv(r"C:\Users\HP\Downloads\breastCancer.csv")
    print("Uploaded breastCancer.csv loaded.")
    print(f"Uploaded CSV dataset shape: {df_csv.shape}")
    print("Original CSV columns (example):", df_csv.columns.tolist()[:5], "...")

    # 1. Drop 'id' column if it exists (common in these CSVs)
    if 'id' in df_csv.columns:
        df_csv = df_csv.drop('id', axis=1)
        print("Dropped 'id' column from CSV dataset.")

    # 2. Drop any 'Unnamed:' columns (also common in these CSVs)
    df_csv = df_csv.loc[:, ~df_csv.columns.str.contains('^Unnamed')]
    if any(col.startswith('Unnamed:') for col in df_csv.columns):
        print("Dropped 'Unnamed:' columns from CSV dataset.")

    # 3. Handle 'diagnosis' column (map 'M' to 1, 'B' to 0)
    if 'diagnosis' in df_csv.columns:
        df_csv['diagnosis'] = df_csv['diagnosis'].map({'M': 1, 'B': 0})
        print("CSV 'diagnosis' column mapped (1=M, 0=B).")
    else:
        raise ValueError("'diagnosis' column not found in breastCancer.csv. Cannot proceed.")

    # 4. Create explicit mapping for the 30 feature columns expected in the CSV
    csv_feature_names_in_order = [
        'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean',
        'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean',
        'fractal_dimension_mean',
        'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
        'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
        'fractal_dimension_se',
        'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst',
        'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst',
        'fractal_dimension_worst'
    ]

    sklearn_feature_names_target = list(cancer_sk.feature_names)

    if len(df_csv.columns) - 1 != len(sklearn_feature_names_target): # -1 for 'diagnosis'
        raise ValueError(
            f"Feature count mismatch after dropping 'id'/'Unnamed' and handling 'diagnosis': "
            f"CSV has {len(df_csv.columns) - 1} features, "
            f"but scikit-learn has {len(sklearn_feature_names_target)}. "
            f"Please double-check the `breastCancer.csv` file's feature columns."
        )

    # Create the renaming dictionary
    current_csv_features = [col for col in df_csv.columns if col != 'diagnosis']
    rename_map = {
        current_csv_features[i]: sklearn_feature_names_target[i]
        for i in range(len(sklearn_feature_names_target))
    }

    df_csv = df_csv.rename(columns=rename_map)
    print("CSV feature columns renamed to match scikit-learn's feature_names.")

    # 5. Ensure df_csv has only the desired common columns in the correct order
    common_cols_final = sklearn_feature_names_target + ['diagnosis']
    df_csv = df_csv[common_cols_final]
    print("CSV dataset columns aligned to scikit-learn standard.")
    print("Renamed CSV columns (example):", df_csv.columns.tolist()[:5], "...")

except FileNotFoundError:
    print("Error: 'breastCancer.csv' not found. Cannot combine datasets.", file=sys.stderr)
    print("Proceeding with only scikit-learn dataset for demonstration.", file=sys.stderr)
    df_combined = df_sk # Use only sklearn data if CSV is missing
except Exception as e:
    print(f"An error occurred loading or processing breastCancer.csv: {e}", file=sys.stderr)
    print("Proceeding with only scikit-learn dataset for demonstration.", file=sys.stderr)
    df_combined = df_sk


# --- 2. Combine Datasets ---
df_sk.columns = [str(col) for col in df_sk.columns] # Ensure consistent string type for columns
if 'df_csv' in locals() and not df_csv.empty:
    df_csv.columns = [str(col) for col in df_csv.columns]
    if all(col in df_csv.columns for col in df_sk.columns):
        print("\nCombining datasets...")
        df_combined = pd.concat([df_sk, df_csv], ignore_index=True)
        print("Datasets combined.")
    else:
        print("\nSkipping dataset combination as breastCancer.csv was loaded, but columns did not align after processing.")
        df_combined = df_sk
else:
    print("\nSkipping dataset combination as breastCancer.csv was not loaded/processed correctly.")
    df_combined = df_sk


print(f"Combined dataset shape: {df_combined.shape}")

# --- 3. Remove Duplicates ---
initial_rows = df_combined.shape[0]
df_combined.drop_duplicates(inplace=True)
rows_after_dedup = df_combined.shape[0]
print(f"Removed {initial_rows - rows_after_dedup} duplicate rows.")
print(f"Combined dataset shape after deduplication: {df_combined.shape}")

# --- 4. Data Preprocessing (Features and Target Split, Scaling) ---

X = df_combined.drop('diagnosis', axis=1)
y = df_combined['diagnosis']

# Get medians of features from the original (unscaled) X for setting default input values in the app
feature_medians = {str(col): X[col].median() for col in X.columns} # Ensure keys are strings

# Imputation (if any missing values are found in the combined dataset)
imputer = SimpleImputer(strategy='median')
numerical_cols = X.select_dtypes(include=np.number).columns
X[numerical_cols] = imputer.fit_transform(X[numerical_cols])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_processed = pd.DataFrame(X_scaled, columns=X.columns)
print("\nFeatures processed (scaled).")


# --- 5. Split Data into Training and Test Sets ---
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)

print(f"\nShape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")


# --- 6. Model Building with Hyperparameter Tuning (GridSearchCV) ---
print("\n--- PERFORMING HYPERPARAMETER TUNING (GridSearchCV) FOR BREAST CANCER MODEL ---")

xgb_model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss', use_label_encoder=False, random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.9],
    'colsample_bytree': [0.7, 0.9]
}

grid_search_xgb_bc = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search_xgb_bc.fit(X_train, y_train)

best_xgb_model_bc = grid_search_xgb_bc.best_estimator_
best_params_bc = grid_search_xgb_bc.best_params_

print("\n--- Breast Cancer Hyperparameter Tuning Complete ---")
print(f"Best AUC-ROC Score during CV: {grid_search_xgb_bc.best_score_:.4f}")
print(f"Best Parameters Found: {best_params_bc}")


# --- 7. Model Evaluation using the BEST Model ---
print("\n--- EVALUATING THE BEST BREAST CANCER MODEL ---")

y_pred_bc = best_xgb_model_bc.predict(X_test)
y_prob_bc = best_xgb_model_bc.predict_proba(X_test)[:, 1]

accuracy_bc = accuracy_score(y_test, y_pred_bc)
precision_bc = precision_score(y_test, y_pred_bc)
recall_bc = recall_score(y_test, y_pred_bc)
f1_bc = f1_score(y_test, y_pred_bc)
roc_auc_bc = roc_auc_score(y_test, y_prob_bc)
conf_matrix_bc = confusion_matrix(y_test, y_pred_bc)
fpr_bc, tpr_bc, thresholds_bc = roc_curve(y_test, y_prob_bc)

print(f"Accuracy: {accuracy_bc:.4f}")
print(f"Precision: {precision_bc:.4f}")
print(f"Recall: {recall_bc:.4f}")
print(f"F1-Score: {f1_bc:.4f}")
print(f"AUC-ROC: {roc_auc_bc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_bc))

print("\nConfusion Matrix:")
print(conf_matrix_bc)

# --- 8. Save Best Model and Preprocessing Objects as a SINGLE BUNDLE ---
print("\n--- SAVING BREAST CANCER MODEL BUNDLE ---")

save_dir_bc = 'breast_cancer_model_assets'
os.makedirs(save_dir_bc, exist_ok=True)

# Collect all components into a single dictionary
breast_cancer_model_bundle = {
    'model': best_xgb_model_bc,
    'scaler': scaler,
    'imputer': imputer,
    'metrics': {
        'accuracy': accuracy_bc,
        'precision': precision_bc,
        'recall': recall_bc,
        'f1_score': f1_bc,
        'roc_auc': roc_auc_bc,
        'conf_matrix': conf_matrix_bc.tolist(),
        'fpr': fpr_bc.tolist(),
        'tpr': tpr_bc.tolist(),
        'feature_names': list(X.columns),
        'feature_medians': feature_medians
    }
}
joblib.dump(breast_cancer_model_bundle, os.path.join(save_dir_bc, 'breast_cancer_model_bundle.pkl'))



print(f"Breast cancer model bundle saved to '{os.path.join(save_dir_bc, 'breast_cancer_model_bundle.pkl')}'")
print("Breast cancer model building, tuning, and evaluation complete!")

# --- Plots (unchanged, as they are for visual evaluation during training) ---
# Visualize Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_bc, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Benign (0)', 'Predicted Malignant (1)'],
            yticklabels=['Actual Benign (0)', 'Actual Malignant (1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Breast Cancer Prediction (Tuned Model)')
plt.show()

# Plot ROC Curve
plt.figure(figsize=(7, 6))
plt.plot(fpr_bc, tpr_bc, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_bc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve (Tuned Model)')
plt.legend(loc="lower right")
plt.show()
```
